```python
import requests
from requests import ConnectionError,HTTPError,Timeout

try:
    res = requests.get("http://www.pythonscraping.com/pages/page1.html")    
    res.raise_for_status()    
except ConnectionError:
    print('找不到伺服器')
except HTTPError:
    print('網頁找不到')
except Timeout:
    print('超過時間沒有回應')
else:
    print('沒有發生問題')
```
