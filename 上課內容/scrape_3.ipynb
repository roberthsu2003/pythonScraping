```python
import requests
response = requests.get('http://pythonscraping.com/pages/warandpeace.html')
```

```python
from bs4 import BeautifulSoup
import re
bs = BeautifulSoup(response.text,'lxml')
bs.__class__
nameList = bs.find_all(name='span',attrs={'class':'green'})
pattern = re.compile('.*Anna.*')

for nameTag in nameList:
    #print(nameTag.__class__)
    #print(nameTag.string)
    if pattern.match(nameTag.string):
        print(nameTag.string)
```

```python
nameList = bs.find_all('span',{'class':{'red','green'}})
pattern = re.compile('.*Anna.*')

for nameTag in nameList:
    print(nameTag.__class__)
    print(nameTag.string)
    if nameTag.string:
        if pattern.match(nameTag.string):
            print(nameTag.string)
```

```python
nameList = bs.find_all(['h1','h2'])


for nameTag in nameList:
    print(nameTag.__class__)
    print(nameTag.string)
    
```

```python
nameList = bs.find_all('span',class_='red')


for nameTag in nameList:
    print(nameTag.__class__)
    print(nameTag.string)
    
```

```python
nameList = bs.find_all(text='the prince')

print(len(nameList))
for nameTag in nameList:
    print(nameTag.__class__)
    print(nameTag.string)
    
```

```python
pattern=re.compile('.*Anna.*')
nameList = bs.find_all(text=pattern)

print(len(nameList))
for nameTag in nameList:
    print(nameTag.__class__)
    print(nameTag.string)
    
```
