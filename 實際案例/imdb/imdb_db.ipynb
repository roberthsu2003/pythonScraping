```python
from bs4 import BeautifulSoup
from selenium import webdriver 
import time
driver = webdriver.Chrome('/Users/roberthsu2003/Downloads/chromedriver')
driver.get('https://www.imdb.com/what-to-watch/fan-favorites/?ref_=hm_fanfav_sm')
time.sleep(5)
soup = BeautifulSoup(driver.page_source,'html.parser')
driver.close()
tabpanelDatas = soup.find(name='div',attrs={'role':'tabpanel'}).find_all('div',{'class':'ipc-poster-card'})
```

```python
for item in tabpanelDatas:
    print(item.get_text())
    print("==========")
```

```python
#取得影片圖片名稱和圖片網址
import os
from urllib.parse import urlparse

for item in tabpanelDatas:
    imageURL = item.find('img',{'class':'ipc-image'})['src']
    print(imageURL)
    a = urlparse(imageURL)
    print(a.path)
    imageName = os.path.basename(a.path)
    print(imageName)
   
```

```python
#取得影片介紹網址
for item in tabpanelDatas:
    detailPath = item.find('a',{'class':'ipc-lockup-overlay'})['href']
    detailUrl = 'https://www.imdb.com' + detailPath
    print(detailUrl)
```

```python
#取得評分
for item in tabpanelDatas:
    rating = item.find('span',{'class':'ipc-rating-star'}).get_text()    
    print(rating)
```

```python
#取得影片名稱
for item in tabpanelDatas:
    title = item.find('span',{'data-testid':'title'}).string    
    print(title)
```

```python
#檢查是否有表格，沒有表格建立新表格
import sqlite3
conn = sqlite3.connect('IMDb.db')
c = conn.cursor()
c.execute(''' SELECT count(name) FROM sqlite_master WHERE type='table' AND name='喜愛影片' ''')
if not c.fetchone()[0]==1:
    print('表格沒有出現')
    #建立表格
    c.execute('''
CREATE TABLE "喜愛影片" (
	"時間"	TEXT NOT NULL,
	"片名"	TEXT NOT NULL,
	"評分"	REAL NOT NULL,
	"介紹網址"	TEXT,
	"圖片下載網址"	TEXT,
	"圖片名稱"	TEXT
)
''')
conn.commit()
conn.close()
```

```python
#新增資料
import sqlite3
conn = sqlite3.connect('IMDb.db')
c = conn.cursor()

for item in tabpanelDatas:
    title = item.find('span',{'data-testid':'title'}).string    
    print("片名:",title)
    rating = item.find('span',{'class':'ipc-rating-star'}).get_text()    
    print("評分:",rating)
    detailPath = item.find('a',{'class':'ipc-lockup-overlay'})['href']
    detailUrl = 'https://www.imdb.com' + detailPath
    print('介紹網址:',detailUrl)
    imageURL = item.find('img',{'class':'ipc-image'})['src']
    print("圖片下載網址:",imageURL)
    a = urlparse(imageURL)
    imageName = os.path.basename(a.path)
    print("圖片名稱:",imageName)
    print("==============")
    c.execute("INSERT INTO 喜愛影片(時間,片名,評分,介紹網址,圖片下載網址,圖片名稱) VALUES (datetime('now','localtime'),?,?,?,?,?)",[title,rating,detailUrl,imageURL,imageName])
conn.commit()
conn.close()
```
