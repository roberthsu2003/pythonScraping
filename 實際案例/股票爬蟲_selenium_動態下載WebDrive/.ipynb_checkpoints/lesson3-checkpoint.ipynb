```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

options = Options()
options.page_load_strategy = 'normal'
id=2330
url = f'https://mis.twse.com.tw/stock/fibest.jsp?stock={id}'
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=options)
driver.get(url)
driver.implicitly_wait(5)
htmlSource = driver.page_source
print(htmlSource)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(htmlSource)

n_tag=soup.find(id=f'{id}_n')
print(f'{n_tag.get_text()}(整股交易行情)')

z_tag=soup.find(id=f'{id}_z')
print(f'成交價:{z_tag.get_text()}')


diff_tag=soup.find(id=f'{id}_diff')
print(f'漲跌:{diff_tag.get_text()}')


tv_tag=soup.find(id=f'{id}_tv')
print(f'成交量:{tv_tag.get_text()}')
```

```python
from selenium.webdriver.common.by import By
element = driver.find_element(By.ID, 'btnChangeToOdd')
element.click()
clickSource = driver.page_source
soup = BeautifulSoup(htmlSource)

z_odd_tag=soup.find(id=f'{id}_z_odd')
print(f'零股成交價:{z_odd_tag.get_text()}')
driver.quit()
```
